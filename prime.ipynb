```python
#!/usr/bin/env python3
# Copyright 2009-2017 BHG http://bw.org/

#checks if a number is prime or not

def isprime(n):
    if n <= 1:
        return False
    for x in range(2, n):
        if n % x == 0:
            return False
    else:
        return True

n = 5
if isprime(n):
    print(f'{n} is prime')
else:
    print(f'{n} not prime')
```

```python
#!/usr/bin/env python3
# Copyright 2009-2017 BHG http://bw.org/

#Generates a list of prime numbers depending on the defined range

def isprime(n):
    if n <= 1:
        return False
    for x in range(2, n):
        if n % x == 0:
            return False
    else:
        return True

def list_primes():
    for n in range(100):
        if isprime(n):
            print(n, end=' ', flush=True)
        print()

list_primes()
```
